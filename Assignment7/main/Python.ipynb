```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import time

start_time = time.perf_counter()

file_path = "C:/Assignment2/dataset/anscombes.csv"
data = pd.read_csv(file_path)

# Segmenting the data by 'dataset'
unique_datasets = data['dataset'].unique()

# Plot settings
plt.figure(figsize=(12, 8))

# Creating scatter plots and regression lines for each dataset
for i, dataset in enumerate(unique_datasets, 1):
    subset = data[data['dataset'] == dataset]
    X = subset[['x']]
    y = subset['y']

    # Create and fit the model
    model = LinearRegression()
    model.fit(X, y)
    y_pred = model.predict(X)

    # Subplot for each dataset
    plt.subplot(2, 2, i)
    plt.scatter(X, y, label='Data Points')
    plt.plot(X, y_pred, color='red', label='Regression Line')
    plt.title(f'Dataset {dataset}')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()

plt.tight_layout()
plt.show()

end_time = time.perf_counter()

execution_time = end_time - start_time
new_execution_time = execution_time*100
print(f"The execution time is {execution_time} seconds")
```
